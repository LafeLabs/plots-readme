```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import jv, yv, kv, iv, hankel1, hankel2
import os
# Define the range of x-values
x = np.linspace(0.1, 15, 500)  # Start from 0.1 to avoid singularity at 0 for some functions
n_values = [0, 1, 2, 3]  # Orders of the Bessel functions to plot

# Create subplots
fig, axs = plt.subplots(2, 3, figsize=(18, 10))
fig.suptitle('Bessel Functions', fontsize=16)

# 1. Bessel function of the first kind, J_n(x)
for n in n_values:
    axs[0, 0].plot(x, jv(n, x), label=f'$J_{n}(x)$, n={n}')
axs[0, 0].set_title('Bessel Function of the First Kind, $J_n(x)$')
axs[0, 0].set_xlabel('x')
axs[0, 0].set_ylabel('$J_n(x)$')
axs[0, 0].legend()
axs[0, 0].grid(True)

# 2. Bessel function of the second kind, Y_n(x)
for n in n_values:
    axs[0, 1].plot(x, yv(n, x), label=f'$Y_{n}(x)$, n={n}')
axs[0, 1].set_title('Bessel Function of the Second Kind, $Y_n(x)$')
axs[0, 1].set_xlabel('x')
axs[0, 1].set_ylabel('$Y_n(x)$')
axs[0, 1].legend()
axs[0, 1].grid(True)

# 3. Modified Bessel function of the first kind, I_n(x)
for n in n_values:
    axs[0, 2].plot(x, iv(n, x), label=f'$I_{n}(x)$, n={n}')
axs[0, 2].set_title('Modified Bessel Function of the First Kind, $I_n(x)$')
axs[0, 2].set_xlabel('x')
axs[0, 2].set_ylabel('$I_n(x)$')
axs[0, 2].legend()
axs[0, 2].grid(True)

# 4. Modified Bessel function of the second kind, K_n(x)
for n in n_values:
    axs[1, 0].plot(x, kv(n, x), label=f'$K_{n}(x)$, n={n}')
axs[1, 0].set_title('Modified Bessel Function of the Second Kind, $K_n(x)$')
axs[1, 0].set_xlabel('x')
axs[1, 0].set_ylabel('$K_n(x)$')
axs[1, 0].legend()
axs[1, 0].grid(True)

# 5. Hankel function of the first kind, H_n^(1)(x)
for n in n_values:
    axs[1, 1].plot(x, hankel1(n, x).real, label=f'$H_{n}^{(1)}(x)$ Real, n={n}')
    axs[1, 1].plot(x, hankel1(n, x).imag, label=f'$H_{n}^{(1)}(x)$ Imag, n={n}', linestyle='--')
axs[1, 1].set_title('Hankel Function of the First Kind, $H_n^{(1)}(x)$')
axs[1, 1].set_xlabel('x')
axs[1, 1].set_ylabel('$H_n^{(1)}(x)$')
axs[1, 1].legend()
axs[1, 1].grid(True)

# 6. Hankel function of the second kind, H_n^(2)(x)
for n in n_values:
    axs[1, 2].plot(x, hankel2(n, x).real, label=f'$H_{n}^{(2)}(x)$ Real, n={n}')
    axs[1, 2].plot(x, hankel2(n, x).imag, label=f'$H_{n}^{(2)}(x)$ Imag, n={n}', linestyle='--')
axs[1, 2].set_title('Hankel Function of the Second Kind, $H_n^{(2)}(x)$')
axs[1, 2].set_xlabel('x')
axs[1, 2].set_ylabel('$H_n^{(2)}(x)$')
axs[1, 2].legend()
axs[1, 2].grid(True)

plt.savefig('plots/plot1.png', transparent=True)
plt.tight_layout(rect=[0, 0.03, 1, 0.95])  # Adjust layout to fit the title
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Create a figure to hold all the plots.  We'll use subplots.
fig = plt.figure(figsize=(12, 18))  # Adjust figure size as needed.

# 1. Sine Function
plt.subplot(6, 2, 1)  # 6 rows, 2 columns, plot 1
x = np.linspace(-2 * np.pi, 2 * np.pi, 400)
y = np.sin(x)
plt.plot(x, y)
plt.title('Sine Function')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.grid(True)

# 2. Cosine Function
plt.subplot(6, 2, 2)  # 6 rows, 2 columns, plot 2
y = np.cos(x)
plt.plot(x, y)
plt.title('Cosine Function')
plt.xlabel('x')
plt.ylabel('cos(x)')
plt.grid(True)

# 3. Tangent Function
plt.subplot(6, 2, 3)  # 6 rows, 2 columns, plot 3
y = np.tan(x)
plt.plot(x, y)
plt.title('Tangent Function')
plt.xlabel('x')
plt.ylabel('tan(x)')
plt.ylim(-10, 10)  # Limit y-axis for better visualization
plt.grid(True)

# 4. Arcsine Function
plt.subplot(6, 2, 4)  # 6 rows, 2 columns, plot 4
x = np.linspace(-1, 1, 400)
y = np.arcsin(x)
plt.plot(x, y)
plt.title('Arcsine Function')
plt.xlabel('x')
plt.ylabel('arcsin(x)')
plt.grid(True)

# 5. Arccosine Function
plt.subplot(6, 2, 5)  # 6 rows, 2 columns, plot 5
y = np.arccos(x)
plt.plot(x, y)
plt.title('Arccosine Function')
plt.xlabel('x')
plt.ylabel('arccos(x)')
plt.grid(True)

# 6. Arctangent Function
plt.subplot(6, 2, 6)  # 6 rows, 2 columns, plot 6
x = np.linspace(-5, 5, 400)
y = np.arctan(x)
plt.plot(x, y)
plt.title('Arctangent Function')
plt.xlabel('x')
plt.ylabel('arctan(x)')
plt.grid(True)

# 7. Hyperbolic Sine Function
plt.subplot(6, 2, 7)  # 6 rows, 2 columns, plot 7
x = np.linspace(-3, 3, 400)
y = np.sinh(x)
plt.plot(x, y)
plt.title('Hyperbolic Sine Function')
plt.xlabel('x')
plt.ylabel('sinh(x)')
plt.grid(True)

# 8. Hyperbolic Cosine Function
plt.subplot(6, 2, 8)  # 6 rows, 2 columns, plot 8
y = np.cosh(x)
plt.plot(x, y)
plt.title('Hyperbolic Cosine Function')
plt.xlabel('x')
plt.ylabel('cosh(x)')
plt.grid(True)

# 9. Hyperbolic Tangent Function
plt.subplot(6, 2, 9)  # 6 rows, 2 columns, plot 9
y = np.tanh(x)
plt.plot(x, y)
plt.title('Hyperbolic Tangent Function')
plt.xlabel('x')
plt.ylabel('tanh(x)')
plt.grid(True)

# 10. Hyperbolic Cotangent Function
plt.subplot(6, 2, 10)  # 6 rows, 2 columns, plot 10
x = np.linspace(-3, 3, 400)
y = 1 / np.tanh(x)  # np.coth is available but can give errors near zero
plt.plot(x, y)
plt.title('Hyperbolic Cotangent Function')
plt.xlabel('x')
plt.ylabel('coth(x)')
plt.ylim(-10, 10)
plt.grid(True)

# 11. Exponential Decay
plt.subplot(6, 2, 11)  # 6 rows, 2 columns, plot 11
x = np.linspace(0, 5, 400)
y = np.exp(-x)  # Exponential decay: e^-x
plt.plot(x, y)
plt.title('Exponential Decay')
plt.xlabel('x')
plt.ylabel('e^(-x)')
plt.grid(True)
plt.savefig('plots/plot2.png', transparent=True)
plt.tight_layout()  # Adjust layout to prevent overlapping titles/labels
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate 1000 evenly spaced points between -4*pi and 4*pi
x = np.linspace(-4 * np.pi, 4 * np.pi, 1000)

# Calculate the sine and cosine values for each x-coordinate
y_sin = np.sin(x)
y_cos = np.cos(x)

# Create a new figure and an axes.
fig, ax = plt.subplots()

# Plot the sine wave in red with a thick line
ax.plot(x, y_sin, color='red', linewidth=2.5, label='sin(x)')

# Plot the cosine wave in blue with a thick line
ax.plot(x, y_cos, color='blue', linewidth=2.5, label='cos(x)')

# Add a title to the plot
ax.set_title('Sine and Cosine Waves (4 Periods)')

# Add x and y axis labels
ax.set_xlabel('x')
ax.set_ylabel('y')

# Add a legend to the plot, loc specifies the location.
ax.legend(loc='upper right')

# Add grid lines for better readability
ax.grid(True)

# Display the plot
plt.show()
```
